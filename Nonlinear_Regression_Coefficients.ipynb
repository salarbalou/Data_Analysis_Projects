```python
import numpy as np
from scipy.optimize import minimize as mn

from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default = 'plotly_dark'
```

```python
!wget 'https://raw.githubusercontent.com/profteachkids/CHE5136_Fall2021/main/WaterVaporPressure.csv'
```

```python
data = np.loadtxt('WaterVaporPressure.csv', delimiter = ',')
T = data[:, 0]
logP = np.log10(data[:, 1])
```

```python
# Do the linear Regression
x = np.c_[T, -logP, np.ones_like(T)]
A, C, D = np.linalg.inv(x.T@x)@x.T@(T*logP)
B = A*C -D
Tplot = np.linspace(np.min(T), np.max(T), 100)
```

```python
# Do the non linear regression using SSE
def SSE(v):
 A, B, C = v
 return np.sum((A-B/(C+T)-logP)**2)
```

```python
Anlr, Bnlr, Cnlr = mn(SSE, (1. ,0., 0.)).x
```

```python
Figure = make_subplots()
Figure.add_scatter(x= T, y=logP, mode= 'markers', name = 'Actual Data')
Figure.add_scatter(x= Tplot, y=A-B/(Tplot+C), mode= 'lines', name = 'linear Regression')
Figure.add_scatter(x= Tplot, y=Anlr-Bnlr/(Tplot+Cnlr), mode= 'lines', name = 'Nonlinear Regression')
Figure.update_layout(xaxis_title = 'Temperature (Kelvin)', yaxis_title = 'logP', height = 600, width = 800)
```
