```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
```

```python
q = 100. #L/min
cain = 1. #mol/L
Tin = 350. #K
V = 100. #L
rho = 1000. #g/L
Cp = 0.239 #J/(g K)
negHr = 5e4 #J/mol
ER = 8750. #K
k0 = 7.2e10 #1/min
UA = 5e4 #J/(min K)
Tc0 = 300 #K
ca0 = 0.5 #mol/L
T0 = 350 #K
tend=10 #min
```

```python
def rhs(t, z):
    Tc= 307.
    ca, T = z
    k = k0*np.exp(-ER/T)
    dc = q*(cain - ca)/V - k*ca
    dT = (rho*q*Cp*(Tin - T) + UA*(Tc-T)+negHr*V*k*ca)/(rho*V*Cp)
    return dc, dT
```

```python
sol=solve_ivp(rhs, (0,tend), (ca0, T0), method='Radau', dense_output=True).sol
```

```python
t_plot = np.linspace(0,tend, 200)
ca, T = sol(t_plot)
```

```python
fig = make_subplots(rows=1, cols=2)
fig.add_scatter(x=t_plot, y=ca, mode='lines', name='Ca',row=1,col=1)
fig.add_scatter(x=t_plot, y=T,mode='lines', name='T',row=1,col=2)
fig.update_layout(width=800,height=600)
```
